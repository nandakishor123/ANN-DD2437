```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot
from pandas import DataFrame
import numpy as np
```

```python
X, y = make_blobs(n_samples=200, centers=2, n_features=2)
one = np.ones((1,2))
np.concatenate((X,one), axis=0)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

```python
W = np.random.rand(1,2)
```

```python
def calculate_predvalue(input_val, weight):
    weighted_sum = []
    for i in range(len(input_val)):
        temp = np.dot(input_val[i], weight.T)
        weighted_sum.append(temp.tolist())
    return weighted_sum
```

```python
max_iter = 200
error = []
for i in range(max_iter):
    print("Epoch:",i)
    thresh=[]
    weighted_sum = []
    for j in range(len(X)):
        weighted_sum = calculate_predvalue(X,W)
    for k in weighted_sum:
        if k[0] > 4:
            thresh.append(1)
        else:
            thresh.append(0)
    error = thresh - y
    mul = 0.001*np.matmul(error, X)
    print("Gradient:", mul)
    W = W - mul
    print(W)
```
